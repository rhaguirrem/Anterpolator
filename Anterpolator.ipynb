```python
import pandas as pd
import plotly.express as px

df = pd.read_csv(filepath_or_buffer='CNN_Fe_Assays.csv')
```

```python
dfValid=df[df['FE'].notnull()]
```

```python
dfFe=dfValid[["mid_x","mid_y","mid_z","FE"]]
dfFe=dfFe[dfFe["FE"]!="NS"]
dfFe=dfFe[dfFe["FE"]!="PENDIENTE"]
```

```python
fig = px.scatter_3d(dfFe[:500], x='mid_x', y='mid_y', z='mid_z', color='FE')
```
